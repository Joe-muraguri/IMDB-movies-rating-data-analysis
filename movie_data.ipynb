# Classify movies based on Rating (Excellent, Good, Average)

```python
def rating(rating):
    if rating >=7:
        return "Excellent"
    elif rating >=6:
        return "Good"
    else:
        return "Average"
```

```python
df['rating_category'] = df['Rating'].apply(rating)
```

# Unique values from Genre

```python
df['Genre']
```

```python
list1= []

for value in df['Genre']:
    list1.append(value.split(','))

```

```python
list1
```

```python
one_d = []

for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
unique_list = []
for item in one_d:
    if item not in unique_list:
        unique_list.append(item)
```

```python
unique_list
```

```python
len(unique_list)
```
